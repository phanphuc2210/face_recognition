## 1. Extracting features from Face

```python
from imutils import paths
import face_recognition
import cv2
import os
```

```python
# Chuyển vào thư mục đang làm việc
directory = r'D:\LearnPython\Face_Recognition'
os.chdir(directory)
```

```python
# Nhận đường dẫn của mỗi tệp trong thư mục có tên Images
# Images ở đây chứa dữ liệu của tôi (thư mục của những người khác nhau)
imagePaths = list(paths.list_images('Images_H'))
knownEncodings = []

names = ["Den", "NgocTrinh", "SonTung", "ThuyNgan", "TranThanh"]
knownNamesIndex = []
```

```python
print(len(imagePaths))
print(imagePaths)
```

```python
# load the harcaascade in the cascade classifier
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt2.xml')
```

```python
# Lặp qua tất cả các đường dẫn của hình ảnh
for (i, imagePath) in enumerate(imagePaths):
    # Trích xuất tên người từ đường dẫn hình ảnh
    name = imagePath.split(os.path.sep)[-2]
    # load the input image and convert it from BGR (OpenCV ordering)
    # to dlib ordering (RGB)
    image = cv2.imread(imagePath)
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Sử dụng Face_recognition để xác định vị trí mặt
    # faces = faceCascade.detectMultiScale(gray,
    #                                  scaleFactor=1.1,
    #                                  minNeighbors=5,
    #                                  minSize=(60, 60),
    #                                  flags=cv2.CASCADE_SCALE_IMAGE)
    boxes = face_recognition.face_locations(rgb, model='hog')
    # compute the facial embedding for the face
    encodings = face_recognition.face_encodings(rgb, boxes)
    # loop over the encodings
    for encoding in encodings:
        knownEncodings.append(encoding)
        knownNamesIndex.append(names.index(name))
```

```python
print(len(knownEncodings))
print(knownEncodings)
```

```python
print(len(knownNamesIndex))
print(knownNamesIndex)
```

```python
#save emcodings along with their names in dictionary data
data = {"encodings": knownEncodings, "label": knownNamesIndex}
```

```python
#use pickle to save data into a file for later use
import pickle

f = open("face_enc_H", "wb")
f.write(pickle.dumps(data))
f.close()
```

## Getting Data Ready:

```python
import pickle

# load the known faces and embeddings saved in last file
data = pickle.loads(open('face_enc_H', "rb").read())
```

### Split the data into features and labels (Usually X and y)


```python
# Create X (features matrix)
X = data['encodings']

# Create y (lables)
y = data["label"]
```

```python
X[0:5]
```

```python
y[0:10]
```

### Split the data into training and test sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
X_train
```

```python
print(y_train)
```

# RandomForestClassifier

```python
from sklearn.metrics import accuracy_score
# Chose the right model and hyper-parameters
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier() #clf = classifer

# We'll keep the default hyper-parametters
clf.get_params()
```

```python
# Fit the model to the training data

clf.fit(X_train, y_train);
```

```python
# Make a prediction
y_pred = clf.predict(X_test)
y_pred
```

```python
print("Accuracy: %.2f %%" %(100*accuracy_score(y_test, y_pred)))
```

```python
print('y_test: ', y_test)
print('y_pred: ', y_pred)
```
